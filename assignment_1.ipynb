```python
import nltk
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lem=WordNetLemmatizer()
```

```python
punctuations='?:!.,;'
```

```python
nltk.download('wordnet')
```

```python
def lemmatize(text):
    words=nltk.word_tokenize(text)
    for word in words:
        if word in punctuations:
            words.remove(word)
    lol=[]
    for word in words:
        lol.append(lem.lemmatize(word,pos='v'))
    return lol
```

```python
df[:100]["review"].apply(lemmatize)[0]
```
